```python
import pandas as pd
import os
DATA_IN_PATH = 'data_in'
BASE_DIR = os.getcwd()
FILE_NAME = 'ChatBotData.csv'
```

## 문장 전체에 대한 분석

```python
data = pd.read_csv(os.path.join(BASE_DIR, DATA_IN_PATH, FILE_NAME), encoding='utf-8')
```

```python
data.head()
```

```python
sentences = pd.concat([data.Q, data.A], axis=0).tolist()
```

```python
tokenized_sentences = [s.split() for s in sentences]
len_syllable = [len(''.join(s)) for s in tokenized_sentences]
len_clause = [len(s) for s in tokenized_sentences]
```

```python
from konlpy.tag import Okt
okt = Okt()
len_morphs = [len(okt.morphs(s)) for s in sentences]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.hist(len_syllable, bins=50, range=[0,50], alpha=0.5, color='r', label='syllable')
plt.hist(len_clause, bins=50, range=[0,50], alpha=0.5, color='b', label='clause')
plt.hist(len_morphs, bins=50, range=[0,50], alpha=0.5, color='g', label='morphs')
plt.yscale('log')
plt.title('Sentence Length Histogram')
plt.xlabel('Sentence Length')
plt.ylabel('Number of Sentences')
plt.legend()
plt.show()
```

```python
pd.DataFrame({"syllable":len_syllable, "clause":len_clause, "morphs": len_morphs}).describe().T
```

```python
plt.figure(figsize=(12,5))
plt.boxplot([len_syllable, len_clause, len_morphs], labels=['len_syllable', 'len_clause' ,'len_morphs'], showmeans=True)
plt.show()
```

## 질문, 답변 각각에 대한 문장 길이 분포 분석

```python
Q_morph_tokens = [len(okt.morphs(s.replace(' ', ''))) for s in data.Q]
A_morph_tokens = [len(okt.morphs(s.replace(' ', ''))) for s in data.A]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.hist(Q_morph_tokens, bins=50, range=[0,50], alpha=0.5, color='r', label='Q')
plt.hist(A_morph_tokens, bins=50, range=[0,50], alpha=0.5, color='b', label='A')
plt.yscale('log')
plt.title('Sentence Length Histogram')
plt.xlabel('Sentence Length')
plt.ylabel('Number of Sentences')
plt.legend()
plt.show()
```

```python
pd.DataFrame({"Answer":A_morph_tokens, "Question" :Q_morph_tokens}).describe().T
```

```python
plt.figure(figsize=(12,5))
plt.boxplot([A_morph_tokens, Q_morph_tokens], labels=['Answer', 'Question']);
```

```python
# 수차례의 실험을 통해 3분위수 근처에서 형태소 단위의 시퀀스를 25개로 설정하는 것이 문장생성 능력이 가장 좋다는 것을 알게됨.
```

```python
[s[0] for s in [okt.pos(s) for s in data.Q] if s[1] in ['Noun', 'Adjective', 'Verb']]
```

```python
Q_NVA_tokens = [n[0] for n in sum([okt.pos(s.replace(' ', '')) for s in data.Q], []) if n[1] in ['Noun', 'Adverb', 'Adjective']]
A_NVA_tokens = [n[0] for n in sum([okt.pos(s.replace(' ', '')) for s in data.A], []) if n[1] in ['Noun', 'Adverb', 'Adjective']]
```

```python
from wordcloud import WordCloud
Q_wordcloud = WordCloud(font_path=os.path.join(BASE_DIR, DATA_IN_PATH, 'NanumGothic.ttf')).generate(' '.join(Q_NVA_tokens))
plt.figure(figsize=(12,12))
plt.imshow(Q_wordcloud)
plt.axis('off')
plt.show()
```

```python
from wordcloud import WordCloud
A_wordcloud = WordCloud(font_path=os.path.join(BASE_DIR, DATA_IN_PATH, 'NanumGothic.ttf')).generate(' '.join(A_NVA_tokens))
plt.figure(figsize=(12,12))
plt.imshow(A_wordcloud)
plt.axis('off')
plt.show()
```
